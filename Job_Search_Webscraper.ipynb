```python
from selenium.webdriver.chrome.options import Options
from selenium.webdriver.chrome.service import Service as ChromeService 
from webdriver_manager.chrome import ChromeDriverManager

#connect to the website naukri.com
URL = 'https://www.naukri.com/'
headers = {"User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/121.0.0.0 Safari/537.36", "DNT":"1","Connection":"close", "Upgrade-Insecure-Requests":"1"}

#take input for job search
job_to_search = input("Enter the job you want to search:")
driver = webdriver.Chrome(service=ChromeService(ChromeDriverManager().install()))

driver.get(URL)

#finding the search box element
search_box = driver.find_element(By.XPATH,"//input[@placeholder='Enter skills / designations / companies']")
search_box.clear()

# Enter the search query into the search box
search_box.send_keys(job_to_search)

#clicking the search button
search_button = driver.find_element(By.XPATH,"//div[@class='qsbSubmit']")
search_button.click()

# for scraping new window
new_window = driver.window_handles[-1]  # Assumes the new window is the last one opened
driver.switch_to.window(new_window)

# Retrieve the URL of the new page
new_page_url = driver.current_url
pagenation_url = new_page_url.split('?')
print("URL of the new page:", new_page_url)

# this is just to ensure that the page is loaded
time.sleep(5)
new_page_html = driver.page_source
soup1 = BeautifulSoup(new_page_html,"html.parser")

data_dic = {'Companies':[],'Job Title':[],'Ratings':[],'Reviews':[],'Experience':[],'Location':[]}

data_dic = {'Companies':[],'Job Title':[],'Ratings':[],'Reviews':[],'Experience':[],'Location':[]}

job_divs = soup1.find_all('div', attrs={'class':'srp-jobtuple-wrapper'})
print(job_divs[0])
for job_div in job_divs:
    title = getTitle(job_div)
    rating = getRating(job_div)
    company = getCompany(job_div)
    review = getReview(job_div)
    experience = getExperienceNeeded(job_div)
    location = getLocation(job_div)
    data_dic.get('Job Title').append(title)
    data_dic.get('Companies').append(company)
    data_dic.get('Ratings').append(rating)
    data_dic.get('Reviews').append(review)
    data_dic.get('Experience').append(experience)
    data_dic.get('Location').append(location)

print(data_dic)
job_df=pd.DataFrame(data_dic)
job_df

```

```python
#title
def getTitle(soup):
    try:
        title = soup.find('a', attrs={'class':'title'}).text
    except:
        title=''
    return title
```

```python
def getRating(soup):
    try:
        rating = soup.find('a', attrs={'class':'rating'}).find('span', attrs={'class':'main-2'}).text
    except:
        rating=''
    return rating
```

```python
def getCompany(soup):
    try:
        company = soup.find('a', attrs={'class':'comp-name mw-25'}).text
    except:
        company=''
    return company
```

```python
def getReview(soup):
    try:
        review = soup.find('a', attrs={'class':'review ver-line'}).text
    except:
        review=''
    return review
```

```python
def getExperienceNeeded(soup):
    try:
        experience = soup.find('span', attrs={'class':'ni-job-tuple-icon ni-job-tuple-icon-srp-experience exp'}).find('span', attrs={'class':'expwdth'}).text
    except:
        experience=''
    return experience
```

```python
def getLocation(soup):
    try:
        location = soup.find('span', attrs={'class':'ni-job-tuple-icon ni-job-tuple-icon-srp-location loc'}).find('span', attrs={'class':'locWdth'}).text
    except:
        location=''
    return location
```
